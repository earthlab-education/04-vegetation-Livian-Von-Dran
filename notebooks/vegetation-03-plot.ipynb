```python
mean_ndvi_2001_2011 = ndvi_da.sel(date=slice('2001', '2011')).mean('date').NDVI
mean_ndvi_2001_2011.plot()
```

```python
# Compute the difference in NDVI before and after water rights returned
ndvi_diff = ndvi_da.sel(date=slice('2012', '2022')).mean('date').NDVI 
- ndvi_da.sel(date=slice('2001', '2011')).mean('date').NDVI
ndvi_diff
```

```python
(
    ndvi_diff.hvplot(
        x='x', y='y', cmap='PiYG', geo=True,
        title='NDVI Changes at Gila River Community: 2001–2011 vs. 2012–2022'
    )
    *  
    gr_boundary_gdf.hvplot(
        x='x', y='y', geo=True,
        fill_color=None, line_color='black'
    )
)
```
